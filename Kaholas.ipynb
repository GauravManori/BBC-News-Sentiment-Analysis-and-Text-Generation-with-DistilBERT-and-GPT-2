```python

import io
import pandas as pd
df = pd.read_csv(io.BytesIO(data_to_load['bbc_news.csv']), nrows = 1000)
```

```python
df.head(10)
```

The purpose of the task is to derive information present in this dataset. For reducing complexity pubDate,guid and link column won't be used for analysis

```python

df = df.drop(['pubDate','guid','link'],axis=1)
```

```python
df.head(10)
```

Now, we are left with two columns. The title describes the headlines of the news while description provides the content of the headlines. Headlines are often misleading so for that purpose we'll concatenate both title and description.

```python
df['text'] = df['title'] + df['description']
```

```python
df['text']
```

we have used the Hugging Face API by importing the transformers library, which is a Python library developed by Hugging Face for natural language processing tasks. We have used the pipeline() function from this library to perform sentiment analysis and text generation.

The DistilBERT model is a smaller, faster, and lighter version of the original BERT model, which makes it more efficient for sentiment analysis on large datasets. Additionally, the "uncased" version of the model means that the input text is converted to lowercase, which makes it more generalizable and less sensitive to the case of the input text. Finally, the pre-trained model "distilbert-base-uncased" has been fine-tuned on the Stanford Sentiment Treebank dataset, which makes it suitable for sentiment analysis tasks.


```python
from transformers import pipeline
model = pipeline('sentiment-analysis',model = 'distilbert-base-uncased') 
```

```python
df
```

```python
#use the pre-trained model to classify the text
pred = model(df['text'].to_list())
```

```python
pred
```

```python
for p in pred:
    if p['label'] == 'LABEL_1':
        p['label'] = 'negative'
    else:
        p['label'] = 'positive'
```

GPT-2 (Generative Pre-trained Transformer 2) is a state-of-the-art language generation model developed by OpenAI. It is known for its ability to generate high-quality text in a variety of styles and formats, making it a popular choice for natural language generation tasks such as text completion, translation, and summarization.

GPT-2 has been pre-trained on a massive amount of text data, making it capable of generating high-quality text with minimal input. This makes it an ideal choice for text generation tasks where a large amount of text needs to be generated quickly and accurately.

```python
generator = pipeline('text-generation', model='gpt2')
```

```python
for text in df['text']:
    generated_text = generator(text, max_length=100, num_return_sequences=1)
    print(generated_text[0]['generated_text']) 
```
